```python
from ollama import Client
import os
from dotenv import load_dotenv

load_dotenv() 
api_key = os.getenv("OLLAMA_API_KEY")

# Initialize the client with the cloud endpoint and your API key
client = Client(
    host='https://ollama.com',
    headers={'Authorization': api_key}
)

simple_ada_code = """
with Ada.Text_Io;
use Ada.Text_Io;

procedure Hello is
begin
   Put_Line("Hello World!");
end Hello;
"""

response = client.chat(
    model='gpt-oss:120b-cloud', 
    messages=[{'role': 'user', 'content': 'Convert the following code from Ada into C++: ' + simple_ada_code }]
)

print(response['message']['content'])
```
